```python
# Read the CSV file into a DataFrame
data = pd.read_csv('videogames.csv')

# Print the DataFrame
print(data)
```

**creating model for recommnedation system**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(new_data['text'])
tfidf
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
def get_recommendations(asin, top_n=5):
    indices = new_data.index[new_data['asin'] == asin].tolist()[0]
    scores = list(enumerate(cosine_sim[indices]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    top_products = [new_data.iloc[score[0]]['asin'] for score in scores[1:top_n+1]]
    return top_products
```

```python
product_id = 'B00HUWA45W'
recommendations = get_recommendations(product_id, top_n=5)
print(recommendations)
```
